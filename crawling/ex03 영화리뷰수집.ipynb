## 1페이지부터 10페이지까지 수집

```python
review_list = []
for i in tq(range(32,33)):
    res = req.get(f"https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&i[…]riptionAlready=false&isMileageSubscriptionReject=false&page={i}")
    soup = bs(res.text, "lxml")
    ico = soup.select("span.ico_viewer")
    for i in ico:
        i.extract()
    review = soup.select("div.score_reple > p > span")
    for i in review:
        review_list.append(i.text.strip())
```

```python
# 반복문의 진행상황을 체크할 수 있는 라이브러리 (로딩바)
from tqdm import tqdm_notebook as tq
```
