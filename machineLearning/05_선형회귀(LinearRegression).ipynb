### 선형회귀 데이터 로딩(보스턴 주택가격 데이터)

```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = load_boston()
```

```python
data.keys()
```

```python
print(data.DESCR)
```

### 모델 학습

```python
X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,
                                                   random_state=722)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
# 선형회귀모델
from sklearn.linear_model import LinearRegression
```

```python
# 모델 학습
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# 학습된 가중치와 절편 확인
print('가중치 :', model.coef_)
print('절편 :', model.intercept_)
```

```python
# 모델 평가
model.score(X_test,y_test) #R square 점수
```

```python
# 다른 평가지표 사용
from sklearn.metrics import mean_absolute_error # MAE
from sklearn.metrics import mean_squared_error # MSE
```

```python
pre = model.predict(X_test) # 예측값 생성
print('MAE :', mean_absolute_error(y_test,pre))
print('MSE :',mean_squared_error(y_test,pre))
```

### 선형모델의 학습 원리

1. 선형모델은 최적의 가중치(w)와 절편(b)를 선택하기위해 MSE(평균제곱오차)를 이용해 각 가설의 적합성을 판단한다.
2. 최적의 가중치(w)와 절편(b)를 만들기위해서 2가지 방식을 활용한다.
    - 해석적 방법(Ordinary Least Squares) -> LinearRegression 클래스로 구현됨
    - 경사하강법(Gradient Descent Algorithm) -> SGDRegressor 클래스로 구현됨
        - 초기 가중치로 부터 기울기 값을 조금씩 수정해서 찾아가는 방식
        - 기울기가 낮아지는 방향으로 계속 업데이트하는 공식

####  가중치 변화에 따른 MSE변화 그래프를 그려보자

```python
# 성적데이터 생성
data = pd.DataFrame([[2,20], [4,40], [8,80],[9,90]],
                    index=['해도','예호','형','병관'],
                    columns=['공부시간','성적점수'])
```

```python
data
```

##### MSE함수 생성

```python
# 문제, 정답, 가중치를 매개변수로 받는다.
def MSE(X,y,w):
    # 예측값 생성
    y_pre = w * X
    # 실제값과 예측값의 차이의 제곱 평균
    error = (y-y_pre) # 오차
    return (error**2).mean() # 평균제곱오차
```

```python
MSE(data['공부시간'], data['성적점수'], 10) # 결과값이 0이 나와야 한다.
```

```python
MSE(data['공부시간'], data['성적점수'], 15)
```

```python
MSE(data['공부시간'], data['성적점수'], 3)
```

##### 그래프 그리기

```python
# 변화하는 기증치(w) 생성
weights = np.linspace(0,20)
weights
```

```python
mse_list = []
for w in weights:
    mse_list.append(MSE(data['공부시간'], data['성적점수'], w))
mse_list
```

```python
plt.figure(figsize=(7,5))
plt.plot(mse_list, marker='*')
plt.ylabel('MSE')
plt.xlabel('W')
plt.show()
```

#### 경사하강법으로 구현된 SGDRegressor를 사용해 보자.

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd_model = SGDRegressor(eta0=0.1,          # 학습률
                         max_iter=6000,       # 반복횟수(가중치 업데이트 횟수)
                         epsilon= 0.001,      # 최소 오차 기준값
                         n_iter_no_change=15, # 업데이트 시 오차가 변경되지 않는걸 허용하는 횟수
                         verbose=1)           #  
```

```python
sgd_model.fit(data[['공부시간']], data['성적점수'])
```

```python
print('가중치 :', sgd_model.coef_)
print("절편 : ", sgd_model.intercept_)
```

```python
sgd_model.predict([[7],[5]])
```

### 다항회귀

 - 선형모델의 수식에서 입력특성에 2차항이 포함된 경우
 - 고차항이 포함되면 데이터의 특성을 더 fit하게 맞출 수 있어 성능개선의 여지가 있다.

```python
boston = load_boston()
boston.keys()
```

```python
data = pd.DataFrame(boston.data, columns=boston. feature_names)
```

```python
X_train, X_test,y_train,y_test = train_test_split(data, boston.target,
                                                  random_state=726)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X_train['LSTAT'], y_train, label='original')

# 선형모델 학습
LSTAT_model = LinearRegression()
LSTAT_model.fit(X_train[['LSTAT']], y_train)
# 선형모델의 예측결과 확인
pre = LSTAT_model.predict(X_train[['LSTAT']])
plt.scatter(X_train[['LSTAT']], pre, label ='predict')

plt.legend() # 범례
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
plt.show()
```

#### 다항회귀 구현하기

1. 특성확장을 통해서 다항식을 만들어준다.
2. sklearn의 PolynomialFeatures를 이용한다

```python
# 특성확장
X_train['LSTAT x LSTAT'] = X_train['LSTAT'] * X_train['LSTAT']
```

```python
# 모델학습
LSTAT_model2 = LinearRegression()
LSTAT_model2.fit(X_train[['LSTAT','LSTAT x LSTAT']], y_train)
# 모델 예측
pre2 = LSTAT_model2.predict(X_train[['LSTAT','LSTAT x LSTAT']])
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X_train['LSTAT'], y_train, label='original') # 오리진날 데이터
plt.scatter(X_train['LSTAT'], pre2, label='predict2') # 다항회귀 결과

plt.legend()
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
plt.show()
```
