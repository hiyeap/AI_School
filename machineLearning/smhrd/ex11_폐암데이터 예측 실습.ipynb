```python
data = np.loadtxt('./data/ThoraricSurgery.csv', delimiter=',')
```

```python
data = pd.DataFrame(data)
```

- 폴란드 브로츠와프 의과대학에서 2013년 공개한 폐암 수술환자의 수술 전 데이터와 수술 후 생존결과를 기록한 의료기록 데이터
- 18개 컬럼으로 구성, 470개의 데이터, 각 항목은 쉼표(,)로 구분
- 종양 유형, 폐활량, 호흡곤란여부, 고통 정도, 기침, 흡연, 천식여부 등 17가지 환자 상태
- 18번째 컬럼은 수술 후 생존 결과 (<span style="color:green;">1 : 생존, 0 : 사망</span>)


```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.3,
                                                    random_state = 1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier()
```

```python
knn.fit(X_train, y_train)
tree.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
tree.score(X_test, y_test)
```
