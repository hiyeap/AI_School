```python
def im_rotate(img, degree):
    h, w = img.shape[:-1]
    crossLine = int(((w * h + h * w) ** 0.5))
    centerRotatePT = int(w / 2), int(h / 2)
    new_h, new_w = h, w
    rotatefigure = cv2.getRotationMatrix2D(centerRotatePT, degree, 1)
    result = cv2.warpAffine(img, rotatefigure, (new_w, new_h))
    return result
```

```python
import easyocr
import cv2
img = cv2.imread('./data/tryread3.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 좌상단 746.9,292.6 / 우하단 888.2,424.2
# (가로 x, 세로 y) - 배열(행, 열)
img_roi = img[292:424, 746:888]
cv2.imshow('org', img_roi)

result = im_rotate(img_roi, 25)
cv2.imshow('img', result)

plt.xticks([]), plt.yticks([])
plt.imshow(result)
```

```python
reader = easyocr.Reader(['ko', 'en']) 
result = reader.readtext(result) 
print(result)
```
