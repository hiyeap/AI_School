#### 데이터 불러오기

```python
# pd.read_csv()
# 구분자 ";" , delimiter = ";"
data = pd.read_csv('./data/student-mat.csv', delimiter=";")
data.head()
# G3 수학 성적 데이터 : 답
# 공부하는 시간(studytime) 특성 1개 사용
```

```python
# info()
data.info()
```

```python
# 컬럼 옵션 설정
pd.set_option('display.max_columns', None)
```

데이터 분리
- 문제와 답 : X와 y 분리
- 훈련데이터와 테스트데이터

```python
# 공부하는 시간(studytime) 특성 1개 사용
# G3 수학 성적 데이터 : 답
# 문제
X = data['studytime']
# 답
y = data['G3']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 테스트 사이즈 0.3
# random_state = 3 고정
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.3,
                                                    random_state = 3)
# 크기확인
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

#### 신경망 모델 만들기
1. 신경망 구조 설계
2. 신경망 학습 및 방법 설정
3. 학습 및 학습과정 시각화
4. 모델 평가 및 예측

#### 신경망 구조 설계

```python
# Sequential : 신경망의 뼈대를 구축하기 위한 함수
# Dense : 신경망 층(뉴런의 묶음)을 구성하는 함수
# Activation : 활성화 설정하는 함수
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
# 신경망 뼈대를 설정
model = Sequential()

# 입력층 설정
# input_dim: 입력 데이터의 특성 형태를 설정(1개 특성)
# units: 생성할 뉴런의 수 
model.add(Dense(units = 4, input_dim = 1))
model.add(Activation('sigmoid'))

# 출력층 설정
# 결과가 나오는 층
model.add(Dense(units = 1))

# 설계된 신경망의 정보를 요약해서 출력
model.summary()
```

신경망 학습 및 평가 방법 설정

```python
#compile()
model.compile(loss = 'mse', # 손실함수(비용함수) : mse(평균제곱오차)
                 optimizer = 'sgd') # 최적화 함수: 경사하강법(sgd:확률적 경사하강법)
```

신경망 학습

```python
# fit(훈련용 문제, 답, epochs = 300)
# 학습 반복 횟수 설정 : epochs
his = model.fit(X_train,y_train,epochs = 300)
```

모델 평가

```python
model.evaluate(X_test, y_test) # loss확인
```

모델 예측

```python
model.predict(X_test)
```

#### 입력 특성이 2개인 신경망을 구성해보자!

- 최종 출력의 뉴런 개수는 1개 설정

```python
X1 = data[['studytime','traveltime']] # 문제
```

```python
y = data['G3'] # 답
```

```python
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, 
                                                    test_size = 0.3,
                                                    random_state = 3)
```

```python
print(X1_train.shape)
print(y_train.shape)
print(X1_test.shape)
print(y_test.shape)
```

```python
# 1. 신경망 구축 - 뼈대 설계
model1 = Sequential()

# 입력층 + 중간층
# add() : 층을 쌓아줌
# Dense() : 층의 내부 설정
# input_dim: 입력할 특성의 개수
# units: 생성할 뉴런의 수 
model1.add(Dense(units = 16, input_dim = 2))
model1.add(Activation('sigmoid')) # 활성화 함수 연결

# 중간층
model1.add(Dense(units = 8))
model1.add(Activation('sigmoid')) # 활성화 함수

# 출력층 설정
# 회귀 : 연속적인 값을 예측
# 출력층 유닛 1개로 설정
# 결과가 나오는 층
model1.add(Dense(units = 1))

# 설계된 신경망의 정보를 요약해서 출력
model1.summary()
```

```python
# compile
model.compile(loss = 'mse', # 손실함수(비용함수) : mse(평균제곱오차)
                 optimizer = 'sgd') # 최적화 함수: 경사하강법(sgd:확률적 경사하강법)
```
