### plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# pyplot은 matplotlib을 간단하고 쉽게 표현해주는 모듈
# 복잡하고 다양하게 표현하는 pylab도 있으나 pyplot에 비해 많은 정보를 입력해야 함
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y) # 하나의 라인
plt.show()
```

#### 스타일 옵션


####  Line style


```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y, ls=":") # 하나의 라인
plt.show()
```

#### Line width

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y, ls=":", lw="5") # 하나의 라인
plt.show()
```

#### Marker


```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y, ls="-.", lw="2", marker='*') # 하나의 라인
plt.show()
```

#### Line color


```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]
plt.plot(x,y, c="m") # 하나의 라인
plt.show()
```

#### 다음 시각화 그래프와 같이 그려보세요~!



```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls='--', c="g", marker="o")
plt.show()
plt.plot(x,y, ls=':', c="r", marker="D", mfc="b", ms="10")
plt.show()
```

#### 그림처럼 만들어보기



```python
plt.plot(x,y, ls='--', lw="5", marker="o", mfc="r", ms="15", mec="g", mew="5")
plt.show()
```

#### 차트 옵션


```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls='--', lw="5", marker="o", mfc="r", ms="15", mec="g", mew="5")
plt.xlim(-5,10) # x축의 범위
plt.ylim(0,12) # y축의 범위
plt.show()
```

##### 틱(tick) 설정
  - 차트나 그래프에서 축상의 위치 표시 지점은 틱(tick)이라고 함
  - 틱에 쓰여준 숫자 혹은 문자는 틱 라벨(tick label)이라고 함
  - 틱의 위치나 라벨은 matplotlib에서 자동으로 정해주지만 수동으로 설정하고 싶다면
xticks, yticks 명령을 사용

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls='--', lw="5", marker="o", mfc="r", ms="15", mec="g", mew="5")

plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

##### 그리드(grid)
- 그림의 틱 위치를 잘 보여주기 위해 차트 내부에 격자 선을 그려주는 것

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y, ls='--', lw="5", marker="o", mfc="r", ms="15", mec="g", mew="5")
plt.grid()
plt.show()
```

#### 여러 개의 플롯을 한번에 그리기

```python
x = np.arange(4)
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, label='No.1')
plt.plot(y, label='No.2')
plt.plot(z, label='No.3')

# loc : 범례의 위치를 표시(upper, center, lower), (left, center, right)
# prop={'size':5} : 범례의 크기 조절
plt.legend(loc='upper right', prop={'size':5}) # 범례 출력
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
plt.show()

plt.plot(z)
plt.show()

plt.plot(y)
plt.plot(z)
plt.show()
```

#### bar chart : 막대 그래프

```python
x = np.arange(6)
y = [80,85,70,60,50,90]
plt.bar(x,y)
plt.show()
```

##### 색상 종류 사이트
https://matplotlib.org/3.2.1/gallery/color/named_colors.html

```python
# 막대 색상 변경(color)
plt.bar(x,y, color="g")
plt.show()
```

```python
plt.bar(x,y, color=['orange','sandybrown','orangered','lightyellow','orchid','mediumpurple'])
plt.show()
```

```python
# 막대 테두리 색상 설정(edgecolor)
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='m')
plt.show()
```

```python
# 막대 테두리 굵기 설정(inewidth)
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='m', linewidth=3)
plt.show()
```

```python
# 투명도 설정(alpha)
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='m', linewidth=3, alpha=0.5)
plt.ylim(40,100)
plt.show()
```

```python
# x, y축 라벨 및 그래프 타이틀 설정
plt.bar(x, y, color=['orange', 'cyan', 'pink'], edgecolor='m', linewidth=3, alpha=0.5)
plt.ylim(40,100)

plt.xlabel("Number") # x축 라벨
plt.ylabel("Score") # y축 라벨
plt.title("Exam") # 그래프 타이틀

plt.show()
```

#### 사고유형이 차대차 사건 중에 죽거나 다친 사람이 많은 발생지시도를 알아보고 시각화해보자!

- 중요한 키워드들이 들어있는 위치 확인
  - 사고유형_대분류에서 '차대차'인 행만 출력
  - '차대차'인 행 중에서 '발생지시도'와 '사상자수'컬럼 확인

```python
data = pd.read_csv('data/Traffic_Accident_2017.csv', encoding="euc-kr")
```

```python
ac = data[data['사고유형_대분류']=="차대차"]
```

```python
ac = ac[["발생지시도", "사상자수"]]
ac
```

```python
ac['발생지시도'].value_counts()
```

#### 각 발생지시도 별로 사상자수의 합계 구하기

```python
ac2 = ac.groupby("발생지시도").sum()
ac2 = ac2["사상자수"]
```

```python
y = ac2
y.values
```

```python
y.index
```

```python
y= y.sort_values(ascending=False)
```

### 예제) 차vs차 교통사고 사상자 수 시각화하기
- x,y축 값 설정

```python
# x축에 들어갈 bar의 개수는 17개
x = np.arange(len(ac2.index))

# y축은 위에서 구한 발생지시도별 사상자수 데이터를 내림차순 정렬로 넣어줌
y = ac2.sort_values(ascending=False)
```

```python
# 차트 크기 변경
plt.figure(figsize=(13,7))
plt.bar(x,y)
plt.xticks(x, labels=y.index)
plt.title("차vs차 교통사고의 사상자 수")
plt.xlabel("지역")
plt.ylabel("사상자 수")
plt.show()
```

```python
x = np.arange(17)
plt.bar(x, y)
plt.xticks(np.arange(17), labels=['경기','충남','전남','경북','충북','경남','전북', '강원', '서울', '대구', '부산', '광주', '울산', '인천', '대전', '제주', '세종'])
plt.title("차vs차 교통사고의 사상자 수")
plt.xlabel("지역")
plt.ylabel("사상자 수")
plt.show()
```
